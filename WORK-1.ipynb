ONLY INHERITANCE SHOWN HERE

```python
class Person:
    def __init__(self, name):
        self.name = name
    def intro(self):
        print(f"Hello, my name is {self.name}")

class Student(Person):
    def __init__(self, name, student_id):
        super().__init__(name)
        self.student_id = student_id

    def introduce_self(self):
        self.intro()  # Using the intro method from the parent class
        print(f"I am a student with ID: {self.student_id}")


student = Student("Pius", "B33118")
student.introduce_self()
```

SHOWING POLYMORPHISM, INHERITANCE AND OVERRIDING in SHAPES 

```python
class Shape:
    def area(self):
        return 0  # Default implementation, can be overridden

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height


rect = Rectangle(5, 10)
print(f"Area of rectangle: {rect.area()}")
```

CREATING BANK ACCOUNTS USING POLYMORPHISM VIA INHERITANCE

```python
class Account:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("Deposit amount must be positive.")
        self.balance += amount
        print(f"{self.owner} deposited UGX {amount}. New balance: UGX {self.balance}")

    def get_balance(self):
        return self.balance


class SavingsAccount(Account):
    def __init__(self, owner, balance=0, interest_rate=0.05):
        super().__init__(owner, balance)
        self.interest_rate = interest_rate

    def add_interest(self):
        interest = self.balance * self.interest_rate
        self.balance += interest
        print(f"Interest of UGX {interest:.2f} added. New balance: UGX {self.balance:.2f}")


        # Create a savings account
kutosi_savings = SavingsAccount("Kutosi", balance=100000)

# Deposit money
kutosi_savings.deposit(50000)

# Add interest
kutosi_savings.add_interest()

# Print final balance
print(f"Final balance: UGX {kutosi_savings.get_balance():,.2f}")
```

GIRL REPLIES TO 3 BOYS BASING ON THE BEST (POLYMORPHISM VIA INHERITANCE)

```python
class Guy:
    def __init__(self, name):
        self.name = name

    def send_message(self):
        raise NotImplementedError("Each guy must send his own message.")


class Boyfriend(Guy):
    def send_message(self):
        return "Hey babe, hope you're doing great today."


class FriendOne(Guy):
    def send_message(self):
        return "Hi beautiful, your smile lights up the room."


class FriendTwo(Guy):
    def send_message(self):
        return "Hey sunshine, just wanted to say you're amazing!"


class Girl:
    def __init__(self, name):
        self.name = name

    def reply_to_sweetest(self, guys):
        sweetest = None
        highest_score = -1

        for guy in guys:
            message = guy.send_message()
            score = self.evaluate_message(message)
            print(f"{guy.name} sent: \"{message}\" (Score: {score})")
            if score > highest_score:
                highest_score = score
                sweetest = guy

        print(f"\n{self.name} replies first to {sweetest.name} ❤️")

    def evaluate_message(self, message):
        # Simple scoring based on sweet words
        sweet_words = ["beautiful", "smile", "sunshine", "amazing", "lights"]
        return sum(word in message.lower() for word in sweet_words) 
    
    # Create guys
g1 = Boyfriend("Daniel")
g2 = FriendOne("Brian")
g3 = FriendTwo("Elvis")

# Create girl
girl = Girl("Lynn")

# She replies to the sweetest message
girl.reply_to_sweetest([g1, g2, g3])
```

AREA OF A RECTANGLE AND A CIRCLE SHOWING POLYMOPHISM VIA INHERITANCE AND OVERRIDING

```python
import math  # Needed for math.pi

class Shape:
    def area(self):
        raise NotImplementedError("Subclasses must implement area() method")

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

# Create instances
rectangle = Rectangle(5, 10)
circle = Circle(7)

# List of shapes
shapes = [rectangle, circle]

# Loop and print areas
for shape in shapes:
    print(f"Area: {shape.area():.2f}")
```

ABSTRACTION PART OF CODE

```python
from abc import ABC, abstractmethod

# Abstract base class
class Vehicle(ABC):
    def __init__(self, make, model, year):              
        self.make = make
        self.model = model
        self.year = year

    @abstractmethod                                      #
    def display_info(self):                              #
        pass  # Must be implemented by subclasses


# Subclass that implements the abstract method
class Car(Vehicle):
    def __init__(self, make, model, year, num_doors, color):    #Initialize additional attribute
        super().__init__(make, model, year)              # Initialize base class attributes
        self.num_doors = num_doors
        self.color = color                       # Initialize additional attribute

    def display_info(self):                              #
        print(f"Vehicle Type: Car")                      #  
        print(f"Make: {self.make}")
        print(f"Model: {self.model}")
        print(f"Year: {self.year}")
        print(f"Number of Doors: {self.num_doors}")
        print(f"Color: {self.color}")


# Create a Car object and display its details   and remember we cant create object vehicle because it's abstract
my_car = Car("Toyota", "Corolla", 2020, 4, 'Blue')               #
my_car.display_info()
```

ABSTRACT EXAMPLE 2 SELF

```python
from abc import ABC, abstractmethod

# Abstract base class
class Team(ABC):
    def __init__(self, name, Motto, City):
        self.name = name
        self.Motto = Motto
        self.City = City

    @abstractmethod
    def get_team_info(self):
        pass


# Subclass defined OUTSIDE the Team class
class Arsenal(Team):
    def __init__(self, name, Motto, City, Founded_Year):
        super().__init__(name, Motto, City)
        self.Founded_Year = Founded_Year

    def get_team_info(self):
        print("Arsenal Team Information:")
        print(f"Team: {self.name}")
        print(f"Motto: {self.Motto}")
        print(f"City: {self.City}")
        print(f"Founded Year: {self.Founded_Year}")


# Create an Arsenal object and display its details
my_team = Arsenal("Arsenal FC", "Victoria Concordia Crescit", "London", 1886)
my_team.get_team_info()
```
